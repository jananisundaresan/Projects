```python
import numpy as np
import pandas as pd
from statsmodels.stats.power import TTestIndPower
from scipy.stats import ttest_ind
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
samplesize=[]
def sample_size(effect_size,alpha,power):
    
    power_analysis = TTestIndPower()
    sample_size = power_analysis.solve_power(effect_size = effect_size, 
                                         power = power, 
                                         alpha = alpha)
    samplesize.append(sample_size)
    return sample_size

    
print("Number of observations for effect size = 0.8, alpha = 0.05, power=0.5: ",round(sample_size(0.8,0.05,0.5),2))
print("Number of observations for effect size = 0.8, alpha = 0.05, power=0.7: ",round(sample_size(0.8,0.05,0.7),2))
print("Number of observations for effect size = 0.8, alpha = 0.05, power=0.8: ",round(sample_size(0.8,0.05,0.8),2))
print("Number of observations for effect size = 0.8, alpha = 0.05, power=0.9: ",round(sample_size(0.8,0.05,0.9),2))
```

```python
#Inference: We can see as the power level increases, the number of observations required for sample size also increases.
```

```python
print("Number of observations for effect size = 0.5, alpha = 0.05, power=0.8: ",round(sample_size(0.5,0.05,0.8),2))
print("Number of observations for effect size = 0.7, alpha = 0.05, power=0.8: ",round(sample_size(0.7,0.05,0.8),2))
print("Number of observations for effect size = 0.8, alpha = 0.05, power=0.8: ",round(sample_size(0.8,0.05,0.8),2))
print("Number of observations for effect size = 0.9, alpha = 0.05, power=0.8: ",round(sample_size(0.9,0.05,0.8),2))
```

```python
#Inference: We can see as the effect size increases, the number of observations required for sample size decreases.
```

```python
print("Number of observations for effect size = 0.5, alpha = 0.01, power=0.8: ",round(sample_size(0.8,0.01,0.8),2))
print("Number of observations for effect size = 0.7, alpha = 0.02, power=0.8: ",round(sample_size(0.8,0.02,0.8),2))
print("Number of observations for effect size = 0.8, alpha = 0.05, power=0.8: ",round(sample_size(0.8,0.05,0.8),2))
print("Number of observations for effect size = 0.9, alpha = 0.07, power=0.8: ",round(sample_size(0.8,0.07,0.8),2))
```

# Power vs Significance Level

```python
# power vs. number of observations 

fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(2,2,1)
fig = TTestIndPower().plot_power(dep_var='nobs',
                                 nobs= np.arange(2, 100),
                                 effect_size=np.array([0.5,0.7, 0.8,0.9]),
                                 alpha=0.01,
                                 ax=ax, title='Power of t-Test' + '\n' + r'$\alpha = 0.01$')
ax.get_legend().remove()
ax = fig.add_subplot(2,2,2)
fig = TTestIndPower().plot_power(dep_var='nobs',
                                 nobs= np.arange(2, 100),
                                 effect_size=np.array([0.5,0.7, 0.8,0.9]),
                                 alpha=0.02,
                                 ax=ax, title='Power of t-Test' + '\n' + r'$\alpha = 0.02$')
ax.get_legend().remove()
ax = fig.add_subplot(2,2,3)
fig = TTestIndPower().plot_power(dep_var='nobs',
                                 nobs= np.arange(2, 100),
                                 effect_size=np.array([0.5,0.7, 0.8,0.9]),
                                 alpha=0.05,
                                 ax=ax, title='Power of t-Test' + '\n' + r'$\alpha = 0.05$')
ax.get_legend().remove()
ax = fig.add_subplot(2,2,4)
fig = TTestIndPower().plot_power(dep_var='nobs',
                                 nobs= np.arange(2, 100),
                                 effect_size=np.array([0.5,0.7, 0.8,0.9]),
                                 alpha=0.06,
                                 ax=ax, title='Power of t-Test' + '\n' + r'$\alpha = 0.06$')
ax.get_legend().remove()

plt.tight_layout()
```
